```python
#Import Library
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
import numpy as np
```

```python
citra1 = imread(fname="jefri.jpeg")  # Membaca citra "jefri.jpeg" dan menyimpannya di variabel citra1
citra2 = imread(fname="arbani.jpeg")  # Membaca citra "arbani.jpeg" dan menyimpannya di variabel citra2

print('Shape citra 1:', citra1.shape)  # Menampilkan bentuk (shape) citra1
print('Shape citra 2:', citra2.shape)  # Menampilkan bentuk (shape) citra2

fig, axes = plt.subplots(1, 2, figsize=(10, 10))  # Membuat tampilan subplot dengan 1 baris dan 2 kolom, ukuran total 10x10
ax = axes.ravel()  # Meratakan array subplot menjadi satu dimensi

ax[0].imshow(citra1, cmap='gray')  # Menampilkan citra1 pada subplot pertama dengan cmap 'gray'
ax[0].set_title("Citra 1")  # Menetapkan judul subplot pertama sebagai "Citra 1"
ax[1].imshow(citra2, cmap='gray')  # Menampilkan citra2 pada subplot kedua dengan cmap 'gray'
ax[1].set_title("Citra 2")  # Menetapkan judul subplot kedua sebagai "Citra 2"
```

```python
copyCitra1 = citra1.copy().astype(float)  # Membuat salinan citra1 dengan tipe data float
copyCitra2 = citra2.copy().astype(float)  # Membuat salinan citra2 dengan tipe data float

m1, n1 = copyCitra1.shape[:2]  # Mendapatkan jumlah baris (m1) dan kolom (n1) dari copyCitra1
output1 = np.empty([m1, n1])  # Membuat array kosong dengan ukuran yang sama dengan copyCitra1 untuk output1

m2, n2 = copyCitra2.shape[:2]  # Mendapatkan jumlah baris (m2) dan kolom (n2) dari copyCitra2
output2 = np.empty([m2, n2])  # Membuat array kosong dengan ukuran yang sama dengan copyCitra2 untuk output2

print('Shape copy citra 1:', copyCitra1.shape)  # Menampilkan bentuk (shape) copyCitra1
print('Shape output citra 1:', output1.shape)  # Menampilkan bentuk (shape) output1

print('m1:', m1)  # Menampilkan jumlah baris (m1) dari copyCitra1
print('n1:', n1)  # Menampilkan jumlah kolom (n1) dari copyCitra1
print()

print('Shape copy citra 2:', copyCitra2.shape)  # Menampilkan bentuk (shape) copyCitra2
print('Shape output citra 2:', output2.shape)  # Menampilkan bentuk (shape) output2

print('m2:', m2)  # Menampilkan jumlah baris (m2) dari copyCitra2
print('n2:', n2)  # Menampilkan jumlah kolom (n2) dari copyCitra2
print()
```

```python
output1 = np.zeros_like(copyCitra1)  # Inisialisasi output1 sebagai array kosong dengan ukuran yang sama

for baris in range(1, m1-1):
    for kolom in range(1, n1-1):
        a1 = baris
        b1 = kolom
        # Menghitung jumlah pixel sekitar
        jumlah = copyCitra1[a1-1, b1-1] + copyCitra1[a1-1, b1] + copyCitra1[a1-1, b1+1] + \
                 copyCitra1[a1, b1-1] + copyCitra1[a1, b1] + copyCitra1[a1, b1+1] + \
                 copyCitra1[a1+1, b1-1] + copyCitra1[a1+1, b1] + copyCitra1[a1+1, b1+1]
        # Menghitung nilai rata-rata dari pixel sekitar
        output1[a1, b1] = (1/9 * jumlah)
```

```python
output2 = np.zeros_like(copyCitra2)  # Inisialisasi output2 sebagai array kosong dengan ukuran yang sama

for baris1 in range(1, m2-1):
    for kolom1 in range(1, n2-1):
        a1 = baris1
        b1 = kolom1
        # Menghitung jumlah pixel sekitar
        jumlah = copyCitra2[a1-1, b1-1] + copyCitra2[a1-1, b1] + copyCitra2[a1-1, b1+1] + \
                 copyCitra2[a1, b1-1] + copyCitra2[a1, b1] + copyCitra2[a1, b1+1] + \
                 copyCitra2[a1+1, b1-1] + copyCitra2[a1+1, b1] + copyCitra2[a1+1, b1+1]
        # Menghitung nilai rata-rata dari pixel sekitar
        output2[a1, b1] = (1/9 * jumlah)
```

```python
# Membuat layout subplot 2x2 dengan ukuran figure 10x10
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
ax = axes.ravel()

# Menampilkan citra 1 pada subplot pertama
ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Input Citra 1")

# Menampilkan citra 2 pada subplot kedua
ax[1].imshow(citra2, cmap='gray')
ax[1].set_title("Input Citra 2")

# Menampilkan output citra 1 pada subplot ketiga
ax[2].imshow(output1, cmap='gray')
ax[2].set_title("Output Citra 1")

# Menampilkan output citra 2 pada subplot keempat
ax[3].imshow(output2, cmap='gray')
ax[3].set_title("Output Citra 2")
```
