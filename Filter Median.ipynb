```python
#import library
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
import numpy as np
```

```python
# Membaca citra 1 dari file "mobil.tif"
citra1 = imread(fname="jefri.jpeg")

# Membaca citra 2 dari file "boneka2.tif"
citra2 = imread(fname="arbani.jpeg")

# Menampilkan dimensi citra 1
print('Shape citra 1: ', citra1.shape)

# Menampilkan dimensi citra 2
print('Shape citra 2: ', citra2.shape)

# Membuat layout subplot 1x2 dengan ukuran figure 10x10
fig, axes = plt.subplots(1, 2, figsize=(10, 10))
ax = axes.ravel()

# Menampilkan citra 1 pada subplot pertama
ax[0].imshow(citra1, cmap='gray')
ax[0].set_title("Citra 1")

# Menampilkan citra 2 pada subplot kedua
ax[1].imshow(citra2, cmap='gray')
ax[1].set_title("Citra 2")
```

```python
# Membuat salinan dari citra 1 dan citra 2
copyCitra1 = citra1.copy()
copyCitra2 = citra2.copy()

# Mendapatkan dimensi citra 1
m1, n1 = copyCitra1.shape[:2]

# Membuat array kosong untuk output citra 1 dengan ukuran yang sama
output1 = np.empty((m1, n1))

# Mendapatkan dimensi citra 2
m2, n2 = copyCitra2.shape[:2]

# Membuat array kosong untuk output citra 2 dengan ukuran yang sama
output2 = np.empty((m2, n2))

# Menampilkan dimensi copy citra 1
print('Shape copy citra 1:', copyCitra1.shape)

# Menampilkan dimensi output citra 1
print('Shape output citra 1:', output1.shape)

# Menampilkan nilai m1 dan n1
print('m1:', m1)
print('n1:', n1)
print()

# Menampilkan dimensi copy citra 2
print('Shape copy citra 2:', copyCitra2.shape)

# Menampilkan dimensi output citra 2
print('Shape output citra 2:', output2.shape)

# Menampilkan nilai m2 dan n2
print('m2:', m2)
print('n2:', n2)
print()
```

```python
# melakukan looping untuk setiap posisi baris dan kolom pada citra 1
for baris in range(0, m1-1):
    for kolom in range(0, n1-1):
        a1 = baris
        b1 = kolom
        # untuk menyimpan nilai piksel
        dataA = [copyCitra1[a1-1, b1-1], copyCitra1[a1-1, b1], copyCitra1[a1-1, b1+1], \
              copyCitra1[a1, b1-1], copyCitra1[a1, b1], copyCitra1[a1, b1+1], \
              copyCitra1[a1+1, b1-1], copyCitra1[a1+1, b1], copyCitra1[a1+1, b1+1]]
        
        # Urutkan
        for i in range(1, 8): # Looping untuk membandingkan nilai dan mengurutkan elemen" dakam dataA
            for j in range(i, 9):
                # kondisi dimana jika elemen dalam dataA[i] lebih besar dari elemen dataA[j] maka elemennya akan ditukar
                if np.any(dataA[i] > dataA[j]):
                    tmpA = dataA[i]
                    dataA[i] = dataA[j]
                    dataA[j] = tmpA
        
        output1[a1, b1] = np.mean(dataA) # nilai piksel pada posisi (a1,b1)
```

```python
# melakukan looping untuk setiap posisi baris dan kolom pada citra 2
for baris in range(0, m2-1):
    for kolom in range(0, n2-1):
        a1 = baris
        b1 = kolom
        # untuk menyimpan nilai piksel
        dataA = [copyCitra2[a1-1, b1-1], copyCitra2[a1-1, b1], copyCitra2[a1-1, b1+1], \
              copyCitra2[a1, b1-1], copyCitra2[a1, b1], copyCitra2[a1, b1+1], \
              copyCitra2[a1+1, b1-1], copyCitra2[a1+1, b1], copyCitra2[a1+1, b1+1]]
        
        # Urutkan
        for i in range(1, 8): # Looping untuk membandingkan nilai dan mengurutkan elemen" dakam dataA
            for j in range(i, 9):
                # kondisi dimana jika elemen dalam dataA[i] lebih besar dari elemen dataA[j] maka elemennya akan ditukar
                if np.any(dataA[i] > dataA[j]):
                    tmpA = dataA[i]
                    dataA[i] = dataA[j]
                    dataA[j] = tmpA
        
        output2[a1, b1] = np.mean(dataA) # nilai piksel pada posisi (a1,b1)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10)) # membuat sebuah figure dengan ukuran 10x10 dan ukuran axes 2x2
ax = axes.ravel() # untuk mengubah array multidimensi menjadi array satu dimensi

# menampilkan input citra 1
ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Input Citra 1")

# menampilkan input citra 2
ax[1].imshow(citra2, cmap = 'gray')
ax[1].set_title("Input Citra 1")

# menampilkan output citra 1
ax[2].imshow(output1, cmap = 'gray')
ax[2].set_title("Output Citra 1")

# menampilkan output citra 2
ax[3].imshow(output2, cmap = 'gray')
ax[3].set_title("Output Citra 2")
```
